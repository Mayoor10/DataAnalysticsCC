```python
#Removing duplicates except the last value (keeping the latest data)
duplicates = df['First Name'].duplicated(keep='last')
new_df=df[~duplicates]
```

```python
new_df
```

```python
#cleaning the family income column (taking higher range as the final value)
new_df['Income'] = new_df['Family Income'].str.replace('7 Lakh+', '-7 Lakh')
new_df[['Income', 'Upper Limit']] = new_df['Income'].str.split('-', 1, expand=True)
new_df['Income'] = new_df['Upper Limit'].str.extract('(\d+)').astype(int)
new_df = new_df.drop(['Upper Limit','Family Income'],axis=1)
```
